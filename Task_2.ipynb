**11.** Есть следующий алгоритм семплинирования, на вход функции подается вектор из целых положительных чисел и число семплов.

```python
def sample(x, c):
    assert len(x) > 0
    
    s = np.sum(x)
    res = []
    for _ in range(c):
        val = s * np.random.random()
        cur, idx = 0, 0        
        while cur + x[idx] <= val:
            cur += x[idx]
            idx += 1
            
        res.append(idx)
    return res

            
sample([50, 3, 1, 7, 20], 5)    
```

Реализуйте данный алгоритм более оптимальным способом, без использования циклов.

```python
def sample(x, c):
    assert len(x) > 0
    
    s = np.sum(x)
    res = []    
    X = np.repeat([x],c,axis=0)
    idx = np.ones(c,dtype=int)
    Idx = np.repeat([idx],c,axis=0)
    rand = np.random.random(c)
    val = s*rand
    Val = np.repeat([val],c,axis=0).transpose()
    res = Idx.sum(where = np.cumsum(X, axis=1)<=Val, axis=1)
    return res

sample([50, 3, 1, 7, 20], 5)    
```
